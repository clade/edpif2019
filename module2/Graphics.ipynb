* Lean by examples : http://matplotlib.org/gallery.html

```python
%matplotlib inline
# Or %pylab
# Or import matplotlib.pyplot as plt; plt.ion()
import numpy as np
from matplotlib.pyplot import *
```

## Functions : 
* plot, loglog, semilogx, semilogy
* xlabel, ylabel, title
* Latex like
* label and legend
* xlim, ylim
* figure
* subplot
* imshow
* savefig : svg, pdf

```python
x = np.linspace(0, 4, 100)*2*np.pi
y = np.sin(x)

subplot(2, 1, 1)
plot(x, y, 'og')
xlabel('Time (s)')
ylabel('Voltage (V)')
title(r'$E = mc^2 + \frac{\theta^2}{2}$')
xlim(0, 4*2*np.pi)
ylim(-2, 2)
grid()
text(10, 1.5, "Hello")

subplot(2, 1, 2)
plot(x, y**2)
xlim(0, 4*2*np.pi)

savefig('/tmp/graph.pdf')
savefig('/tmp/graph.svg')
```

## Object oriented

```python
from matplotlib.pyplot import figure

f = figure()
ax1, ax2 = f.subplots(2, 1)
#(ax1, ax2), (_,_) = f.subplots(2, 2)



ax1.plot(x, y, 'og')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Voltage (V)')
ax1.set_title(r'$E = mc^2 + \frac{\theta^2}{2}$')
ax1.set_xlim(0, 4*2*np.pi)
ax1.set_ylim(-2, 2)
ax1.grid()
ax1.text(10, 1.5, "Hello")


ax2.plot(x, y**2)
ax2.set_xlim(0, 4*2*np.pi)

f.savefig('/tmp/graph2.pdf')
```
