```python
def fluorescence(omega, omega_0, Gamma):
    return Gamma**2/((omega-omega_0)**2 + Gamma**2)

omega_0 = 0
omega = np.linspace(-20E6, 20E6)*2*np.pi

plot(omega, fluorescence(omega, omega_0, Gamma=2*np.pi*6E6))
```

```python
def my_vectorize(initial_function):
    def vectorized_function(a):
        out = np.zeros(len(a))
        for i in range(len(a)):
            out[i] = initial_function(a[i])
        return out
    return vectorized_function

@my_vectorize
def my_abs(x):
    if x>0:
        return x
    else:
        return -x
#my_abs = my_vectorize(my_abs)

x = np.linspace(-1, 1)
plot(x, my_abs(x))
```

```python
x = np.linspace(-1, 1, 1000000)
%timeit y = my_abs(x)
```
